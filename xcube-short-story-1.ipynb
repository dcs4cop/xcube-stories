Import the xcube CCI store class.

```python
from xcube.core.store import new_data_store
```

Create a CCI store.

```python
cci_store = new_data_store('cciodp')
```

Retrieve land surface temperature data for Europe in March 2017

```python
cube = cci_store.open_data(
    data_id='esacci.LST.mon.L3C.LST.SLSTR.Sentinel-3A.SLSTRA.3-00.DAY',
    variable_names=['lst'],
    bbox=(-5, 35, 25, 65),
    time_range=['2017-03-01', '2017-03-31'],
    opener_id='dataset:zarr:cciodp'
)
```

Plot the temperature on a map.

```python
cube.lst.isel(time=0).plot.imshow()
```
